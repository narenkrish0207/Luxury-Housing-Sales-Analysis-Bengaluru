Data Cleaning

```python
# Remove 'Cr' and convert to float
house_df['Ticket_Price_Cr'] = house_df['Ticket_Price_Cr'].str.replace('Cr', '', regex=False).str.strip().str.replace('Rs','').str.replace('₹','')
house_df['Ticket_Price_Cr'] = pd.to_numeric(house_df['Ticket_Price_Cr'], errors='coerce')
```

```python
house_df['Ticket_Price_Cr']
```

```python
#  Fill missing values

house_df['Unit_Size_Sqft'] = house_df['Unit_Size_Sqft'].fillna(house_df['Unit_Size_Sqft'].mean())
house_df['Ticket_Price_Cr'] = house_df['Ticket_Price_Cr'].fillna(house_df['Ticket_Price_Cr'].mean())
house_df['Amenity_Score'] = house_df['Amenity_Score'].fillna(house_df['Amenity_Score'].median())
house_df['Buyer_Comments'] = house_df['Buyer_Comments'].fillna("No Comments")
```

```python
house_df
```

```python
# Normalize text columns

text_cols = ['Micro_Market', 'Developer_Name', 'Configuration',
             'Transaction_Type', 'Buyer_Type', 'Possession_Status',
             'Sales_Channel', 'NRI_Buyer']
```

```python
for col in text_cols:
    house_df[col] = house_df[col].str.strip().str.title()
```

```python
house_df
```

```python
house_df["Price_per_Sqft"] = house_df["Ticket_Price_Cr"] * 10000000 / house_df["Unit_Size_Sqft"]
```

```python
house_df['Purchase_Quarter'] = pd.to_datetime(house_df['Purchase_Quarter'])
house_df['Quarter_Number'] = house_df['Purchase_Quarter'].dt.quarter
house_df['Quarter_Number']
```

```python
# Create Booking_Flag from Transaction_Type
# Let's assume Primary = booked (1), Secondary = not booked (0)
house_df['Booking_Flag'] = house_df['Transaction_Type'].str.lower().map({'primary': 1, 'secondary': 0})
```
